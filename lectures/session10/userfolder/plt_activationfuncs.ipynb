```python
import numpy as np
import matplotlib.pyplot as plt
```

Sigmoid

```python
x = np.linspace(-5,5,400)
y = 1/(1+np.e**(-x))

plt.plot(x,y)
plt.title("sigmoid function")
plt.show()
```

tanh

```python
x = np.linspace(-3,3,200)
y = np.tanh(x)

plt.plot(x,y)
plt.title("tanh")
plt.show()
```

ReLU

```python
y = (x + np.abs(x))/2

plt.plot(x,y)
plt.title("Rectifier Linear Unit (ReLU)")
plt.show()
```
