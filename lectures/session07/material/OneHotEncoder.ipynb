# One-hot encoding

Find below some snippets for performing a one-hot ecoding using scikit-learn - demonstrated on the example of amino acids.

```python
# define vocabulary for amino acids:

AA_ALPHABET = [
    "A","R","N","D","C","E","Q","G","H","I",
    "L","K","M","F","P","S","T","W","Y","V","X"  # X = unknown/other
]
```

```python

from sklearn.preprocessing import OneHotEncoder
import numpy as np

alphabet = np.array(AA_ALPHABET).reshape(-1, 1) # -1 in numpy tells np to use the lenght of the list
encoder = OneHotEncoder(categories=[AA_ALPHABET], handle_unknown="ignore", sparse_output=False)

# Fit with single-character categories; 
# if no "categories" is specified, fitting will simply learn from the data 
encoder.fit(alphabet)


def one_hot_sequence_sklearn(seq):
    # X will be assigned for each character not matching any other from the alphabeth
    seq = ''.join(c if c in AA_ALPHABET else 'X' for c in seq.upper()) 

    chars = np.array(list(seq.upper())).reshape(-1, 1)
    return encoder.transform(chars)

print("Vocabulary (column order):", encoder.get_feature_names_out())
print(one_hot_sequence_sklearn("MKT"))  # (3, 21)
```

Additionally to the minimal one-hot encoding, physicochemical values can be 

```python
#  Example minimal descriptors (generated values - no scientific value!!)
hydropathy = { # Kyte-Doolittle (approx)
    "A": 1.8,"R": -4.5,"N": -3.5,"D": -3.5,"C": 2.5,"Q": -3.5,"E": -3.5,"G": -0.4,"H": -3.2,"I": 4.5,
    "L": 3.8,"K": -3.9,"M": 1.9,"F": 2.8,"P": -1.6,"S": -0.8,"T": -0.7,"W": -0.9,"Y": -1.3,"V": 4.2,"X": 0.0
}
charge = { # at pH ~7 (very simplified net charge)
    "A":0,"R":1,"N":0,"D":-1,"C":0,"E":-1,"Q":0,"G":0,"H":0.1,"I":0,
    "L":0,"K":1,"M":0,"F":0,"P":0,"S":0,"T":0,"W":0,"Y":0,"V":0,"X":0
}

def one_hot_plus_physchem(seq):
    # X will be assigned for each character not matching any other from the alphabeth
    seq = ''.join(c if c in AA_ALPHABET else 'X' for c in seq.upper()) 

    O = one_hot_sequence_sklearn(seq)  # (L, 21)
    H = np.array([hydropathy.get(r, 0.0) for r in seq.upper()], dtype=np.float32).reshape(-1, 1)
    C = np.array([charge.get(r, 0.0) for r in seq.upper()], dtype=np.float32).reshape(-1, 1)
    return np.concatenate([O, H, C], axis=1)  # (L, 23)

features = one_hot_plus_physchem("MKTDä")
print(features.shape)
```
